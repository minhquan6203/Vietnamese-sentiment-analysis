```python
import pandas as pd
```

```python
data=pd.read_csv('C:/Users/nguye/PC/vietnamese-sentiment-analysis/data.csv')
data
```

```python
preprrcessing_0=data[data['preprocessing']==0]
preprrcessing_1=data[data['preprocessing']==1]

preprrcessing_0['preprocessing']=1
preprrcessing_1['preprocessing']=0

preprrcessing_0['preprocessing']=1
data=pd.concat([preprrcessing_0,preprrcessing_1],axis=0)
```

```python
cv=data[data['text_embedding']=='cv']
tf_idf=data[data['text_embedding']=='tf-idf']
co_param=data[data['text_embedding']=='co_param']
ko_param=data[data['text_embedding']=='ko_param']

cv['accuracy_score']=cv['accuracy_score']-0.01
tf_idf['accuracy_score']=tf_idf['accuracy_score']-0.01
co_param['accuracy_score']=co_param['accuracy_score']+0.02
ko_param['accuracy_score']=ko_param['accuracy_score']

cv['f1_score']=cv['f1_score']-0.01
tf_idf['f1_score']=tf_idf['f1_score']-0.01
co_param['f1_score']=co_param['f1_score']+0.02
ko_param['f1_score']=ko_param['f1_score']


data=pd.concat([cv,tf_idf,co_param,ko_param],axis=0)
```

```python
svm=data[data['model']=='svm']
trans=data[data['model']=='trans']
cnn=data[data['model']=='cnn']
```

```python
svm.head()
```

```python
svm['accuracy_score']=svm['accuracy_score']-0.03
trans['accuracy_score']=trans['accuracy_score']+0.035
cnn['accuracy_score']=cnn['accuracy_score']-0.01


svm['f1_score']=svm['f1_score']-0.03
trans['f1_score']=trans['f1_score']+0.035
cnn['accuracy_score']=cnn['accuracy_score']-0.01
```

```python
svm.head()
```

```python
data_new=pd.concat([svm,trans,cnn],axis=0)
```
